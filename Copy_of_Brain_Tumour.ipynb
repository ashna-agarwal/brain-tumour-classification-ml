

# Literature Review: Brain Tumor Detection Using Machine Learning and Deep Learning

Brain tumor detection is an important area in medical image analysis because early diagnosis significantly improves treatment outcomes. Magnetic Resonance Imaging (MRI) is widely used to visualize brain structures and detect abnormalities such as tumors. However, manual analysis of MRI scans by radiologists can be time-consuming and subjective. Therefore, many researchers have proposed **machine learning (ML) and deep learning (DL) techniques** to automate brain tumor detection and classification. This section reviews eight research papers that explore different AI-based approaches for brain tumor recognition.

---

# 1. MRI-Based Brain Tumor Detection Using Convolutional Neural Networks

This study proposes a **deep learning-based system using Convolutional Neural Networks (CNNs)** to automatically detect brain tumors from MRI images. The goal of the research is to reduce the dependency on manual analysis and improve diagnostic accuracy.

The proposed method first performs **image preprocessing**, which includes resizing MRI images and normalizing pixel values. These preprocessed images are then passed to a CNN architecture consisting of convolutional layers, activation layers, pooling layers, and fully connected layers. The convolutional layers extract important features such as tumor edges, abnormal textures, and intensity differences in the MRI scans.

The model is trained using labeled MRI datasets where images are classified as tumor or non-tumor. During training, the CNN learns tumor patterns through backpropagation and weight optimization. Experimental results show that the CNN-based approach achieves high classification accuracy compared with traditional machine learning techniques.

**Short Literature Review Summary:**
This paper presents a CNN-based approach for automatic brain tumor detection using MRI images. The system preprocesses MRI scans and uses convolutional layers to extract spatial features such as edges and textures. The trained model classifies images as tumor or non-tumor with high accuracy, demonstrating the effectiveness of deep learning in medical image analysis.

---

# 2. MRI-Based Brain Tumor Classification Using Ensemble Deep Features and Machine Learning Classifiers

This research focuses on combining **deep learning feature extraction with machine learning classifiers** to improve brain tumor classification performance. Instead of directly using CNNs for classification, the authors use pretrained deep learning models to extract important image features.

The methodology begins with preprocessing MRI images through resizing and normalization. Next, deep features are extracted using pretrained networks such as **VGG16, ResNet50, and DenseNet**. These networks have been trained on large datasets and are capable of capturing complex image patterns.

The extracted deep features are then combined using **feature ensemble techniques** to create a more informative feature representation. These features are fed into machine learning classifiers including **Support Vector Machine (SVM), Random Forest, and K-Nearest Neighbors (KNN)** for final classification.

Experimental results indicate that the **SVM classifier combined with ensemble deep features provides the highest accuracy**, demonstrating the advantage of combining deep learning and classical ML algorithms.

**Short Literature Review Summary:**
This study proposes a hybrid framework combining deep feature extraction from pretrained CNN models with machine learning classifiers for brain tumor classification. Ensemble deep features extracted from multiple networks improve feature representation and enhance classification accuracy. The results show that SVM with ensemble deep features outperforms other classifiers.

---

# 3. Brain Tumor Detection Using YOLOv7 Deep Learning Model

This research introduces a **YOLOv7-based deep learning approach** for detecting brain tumors in MRI images. Unlike classification models that only identify the presence of tumors, YOLOv7 is designed for **object detection**, meaning it can also identify the tumor location within the MRI scan.

The model processes MRI images through a deep neural network that detects tumor regions using bounding boxes. The YOLOv7 architecture consists of convolutional layers, feature extraction modules, and detection layers that allow the network to simultaneously perform localization and classification.

The dataset used in this study contains MRI images of different tumor types including **glioma, meningioma, and pituitary tumors**. The model is trained using labeled images where tumor regions are marked. During prediction, YOLOv7 identifies the tumor location and classifies the tumor type.

Experimental results show that the YOLOv7 model achieves high detection accuracy and faster processing speed compared to conventional CNN-based classification models.

**Short Literature Review Summary:**
This paper proposes a YOLOv7 deep learning model for brain tumor detection and localization in MRI images. The model detects tumor regions using bounding boxes and classifies tumor types simultaneously. The proposed approach demonstrates high accuracy and real-time detection capabilities.

---

# 4. Brain Tumor Detection Using Hybrid Vision Transformer and GRU

This study introduces a **hybrid deep learning model combining Vision Transformers (ViT) and Gated Recurrent Units (GRU)** for brain tumor detection. Transformers are advanced deep learning models capable of capturing global relationships within images.

In this approach, MRI images are first processed by a **Vision Transformer**, which divides the image into smaller patches and analyzes spatial relationships between them. These extracted features are then passed to a **GRU network**, which helps capture sequential dependencies in the feature representation.

The hybrid architecture improves the model's ability to detect subtle tumor patterns and spatial variations in MRI scans. The results demonstrate that the proposed model achieves higher accuracy compared to conventional CNN-based models.

**Short Literature Review Summary:**
The paper proposes a hybrid deep learning architecture combining Vision Transformers and GRU networks for brain tumor classification. The model captures both spatial and sequential features from MRI images, leading to improved classification performance compared with traditional CNN approaches.

---

# 5. Brain Tumor Detection Using Deep Wavelet Autoencoder

This research proposes a method that integrates **wavelet transform techniques with deep autoencoder networks** for brain tumor detection.

In the proposed approach, MRI images are first transformed using **wavelet decomposition**, which helps extract frequency-based features and reduce noise in the images. These wavelet features are then passed into a **deep autoencoder**, which learns compressed representations of the MRI data.

The autoencoder learns important features of tumor structures while reducing redundant information. These learned features are then used for classification of tumor and non-tumor images.

The experimental results indicate that combining wavelet transform with deep autoencoders improves the detection of subtle tumor patterns in MRI images.

**Short Literature Review Summary:**
This study proposes a brain tumor detection system using wavelet transform and deep autoencoder networks. The wavelet transform extracts frequency-based features from MRI images, while the autoencoder learns compact feature representations. The combined approach improves tumor detection performance.

## **MLDC Comparison Analysis of the Reviewed Papers**

The reviewed literature shows that most studies follow the Machine Learning Development Cycle (MLDC), but the depth of implementation varies among the papers. The MLDC generally consists of the following stages: data collection, data preprocessing, exploratory data analysis (EDA), feature engineering, model training, model evaluation, and deployment considerations.

Across the eight reviewed papers, data collection and preprocessing are consistently implemented, as MRI images typically require normalization, resizing, and noise removal before being used for machine learning models. Almost all papers applied image preprocessing techniques such as normalization, segmentation, or augmentation, which improved the quality of MRI scans and allowed the models to better learn tumor patterns.

However, explicit exploratory data analysis (EDA) was rarely reported. Most of the studies focused directly on preprocessing and model training without presenting detailed statistical analysis or visualization of the datasets. This indicates that many medical image classification studies rely more on deep learning architectures rather than traditional data analysis workflows.

In terms of feature engineering, the approaches differ significantly. Traditional machine learning papers, such as the hybrid ensemble classifier study, relied on handcrafted feature extraction methods like GLCM texture features, wavelet transforms, and HOG descriptors. These techniques require manual feature engineering and selection before classification. In contrast, modern deep learning-based papers utilized automatic feature extraction through convolutional neural networks (CNNs), transfer learning models such as VGG16 and ResNet, and object detection frameworks like YOLO. These models automatically learn hierarchical image features without manual feature design.

Feature transformation and selection techniques were applied in some studies. For example, PCA was used in certain hybrid machine learning approaches to reduce dimensionality and improve classification performance. Similarly, ensemble feature techniques were employed to combine feature representations from multiple deep learning architectures.

All reviewed studies implemented model training and evaluation stages of the MLDC, typically using supervised learning methods with labeled MRI datasets. Evaluation metrics commonly included accuracy, precision, recall, F1-score, sensitivity, specificity, and mean average precision (mAP) for object detection models. Among the reviewed approaches, deep learning architectures such as CNNs, hybrid transformer models, and ensemble feature frameworks demonstrated superior performance compared with traditional machine learning models.

Overall, the literature indicates that modern brain tumor detection research increasingly relies on deep learning models with automated feature extraction, while traditional feature engineering techniques are gradually being replaced. Future research directions suggested in the literature include the use of larger datasets, advanced transformer architectures, and hybrid deep learning frameworks to further improve tumor detection accuracy.

## SECTION 2 — Import Libraries

```python
import numpy as np
import pandas as pd
import cv2
import matplotlib.pyplot as plt
import seaborn as sns
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers, models
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
from tensorflow.keras.utils import to_categorical
from sklearn.metrics import confusion_matrix, classification_report
import os

# Imports for Transfer Learning
from tensorflow.keras.applications import MobileNetV2
from tensorflow.keras.layers import GlobalAveragePooling2D, Dense, Dropout
from tensorflow.keras.models import Model
```

## SECTION 3 — Dataset Loading

This section loads the MRI images from the specified directory. Each image will be resized to 128x128 pixels to standardize input for the CNN and to speed up CPU training. Labels are extracted from folder names.

```python
dataset_path = '/content/brain_tumor_dataset'
IMG_SIZE = 128

images = []
labels = []

# Assuming the structure is dataset_path -> (Testing/Training) -> class_name -> images
# First, get the main categories (Testing, Training)
main_categories = sorted(os.listdir(dataset_path))
all_class_names_set = set() # To collect unique actual class names

for main_cat in main_categories:
    main_cat_path = os.path.join(dataset_path, main_cat)
    if os.path.isdir(main_cat_path):
        # Now, get the actual tumor classes inside Testing/Training
        current_class_names = sorted(os.listdir(main_cat_path))
        for class_name in current_class_names:
            class_path = os.path.join(main_cat_path, class_name)
            if os.path.isdir(class_path): # Ensure it's a directory (e.g., 'notumor', 'glioma')
                all_class_names_set.add(class_name) # Add to the set of all unique class names
                for img_name in os.listdir(class_path):
                    img_path = os.path.join(class_path, img_name)
                    try:
                        img = cv2.imread(img_path)
                        if img is not None:
                            img = cv2.resize(img, (IMG_SIZE, IMG_SIZE))
                            images.append(img)
                            labels.append(class_name)
                        else:
                            print(f"Warning: Could not read image {img_path}")
                    except Exception as e:
                        print(f"Error loading image {img_path}: {e}")

# Convert lists to numpy arrays
images = np.array(images)
labels = np.array(labels)

print(f"Detected actual class folders: {sorted(list(all_class_names_set))}")
print("Dataset loading complete.")

# Define unique_classes here as it's used in subsequent cells
unique_classes = np.unique(labels)
```

```python
print(f"Number of images: {len(images)}")
print(f"Class names: {np.unique(labels)}")
print(f"Dataset shape (images): {images.shape}")
print(f"Dataset shape (labels): {labels.shape}")
```

## SECTION 6 — Label Encoding

This section converts the categorical string labels into numerical format using `LabelEncoder` and then applies one-hot encoding using `to_categorical`. This is necessary for training neural network models.

```python
# Encode string labels to numerical labels
label_encoder = LabelEncoder()
numeric_labels = label_encoder.fit_transform(labels)

# Convert numerical labels to one-hot encoding
one_hot_labels = to_categorical(numeric_labels, num_classes=len(np.unique(labels)))

print(f"Original labels (first 5): {labels[:5]}")
print(f"Numeric labels (first 5): {numeric_labels[:5]}")
print(f"One-hot labels (first 5): {one_hot_labels[:5]}")
print(f"Shape of one-hot labels: {one_hot_labels.shape}")
```

## SECTION 7 — Train Test Split

The dataset is split into training and testing sets. A 80/20 split is used, and stratification is applied to ensure that each class is represented proportionally in both the training and testing sets, which is crucial for imbalanced datasets.

```python
X_train, X_test, y_train, y_test = train_test_split(
    images,
    one_hot_labels,
    test_size=0.2,
    random_state=42,
    stratify=numeric_labels # Stratify by numeric labels to maintain class distribution
)

print(f"Shape of X_train: {X_train.shape}")
print(f"Shape of X_test: {X_test.shape}")
print(f"Shape of y_train: {y_train.shape}")
print(f"Shape of y_test: {y_test.shape}")
```

## SECTION 8 — Data Normalization

Image pixel values are typically in the range of 0-255. Normalizing these values by dividing by 255 scales them to a range of 0-1. This helps in faster convergence of the neural network and often improves performance.

```python
X_train = X_train.astype('float32') / 255.0
X_test = X_test.astype('float32') / 255.0

print(f"Min pixel value in X_train: {np.min(X_train)}")
print(f"Max pixel value in X_train: {np.max(X_train)}")
```

## SECTION 9 — CNN Model

This section defines a simple Convolutional Neural Network (CNN) architecture suitable for CPU training. The model consists of convolutional layers for feature extraction, max-pooling layers for dimensionality reduction, flatten layer, and dense layers for classification with a dropout layer for regularization. The output layer uses a softmax activation for multi-class classification.

```python
# Load the pretrained base model (MobileNetV2) for transfer learning

# Ensure IMG_SIZE and num_output_classes are defined for standalone execution
# These are typically defined in earlier data loading and preprocessing steps.
# If running this cell in isolation or after a kernel restart,
# ensure preceding cells that define these variables have been executed.
# For robustness, we'll define them here as fallback if not already present.
if 'IMG_SIZE' not in globals():
    IMG_SIZE = 128  # Default value as per notebook specification

# The number of unique classes should come from the loaded dataset.
# If 'unique_classes' is not globally defined, we assume 4 classes as per dataset description.
if 'unique_classes' not in globals():
    num_output_classes = 4 # Fallback: glioma, meningioma, notumor, pituitary
else:
    num_output_classes = len(unique_classes)


base_model = MobileNetV2(
    weights='imagenet',
    include_top=False,  # Exclude the classifier layers at the top
    input_shape=(IMG_SIZE, IMG_SIZE, 3)
)

# Freeze the layers of the base model so they are not updated during training
for layer in base_model.layers:
    layer.trainable = False

# Add custom classification layers on top of the base model
x = base_model.output
x = GlobalAveragePooling2D()(x)  # Convert features to a single vector
x = Dense(128, activation='relu')(x)
x = Dropout(0.5)(x)  # Dropout for regularization
outputs = Dense(num_output_classes, activation='softmax')(x)  # Output layer for the derived number of classes

# Create the new model
model = Model(inputs=base_model.input, outputs=outputs)

model.summary()

# Compile the model
model.compile(
    optimizer='adam',
    loss='categorical_crossentropy',
    metrics=['accuracy']
)

print("Transfer Learning model (MobileNetV2) built and compiled.")
```

## SECTION 10 — Model Training

The CNN model is trained using the training data. The training process runs for a specified number of epochs and uses a defined batch size. The training history (loss and accuracy over epochs) is stored for later visualization.

```python
epochs = 8
batch_size = 32

history = model.fit(
    X_train, y_train,
    epochs=epochs,
    batch_size=batch_size,
    validation_data=(X_test, y_test)
)

print("Model training complete.")
```

## SECTION 11 — Training Graphs

This section visualizes the training and validation accuracy and loss over epochs. These plots are crucial for understanding the model's learning process, detecting overfitting or underfitting, and evaluating its generalization ability.

```python
plt.figure(figsize=(12, 5))

# Plot training & validation accuracy values
plt.subplot(1, 2, 1)
plt.plot(history.history['accuracy'])
plt.plot(history.history['val_accuracy'])
plt.title('Model Accuracy')
plt.ylabel('Accuracy')
plt.xlabel('Epoch')
plt.legend(['Train', 'Validation'], loc='upper left')

# Plot training & validation loss values
plt.subplot(1, 2, 2)
plt.plot(history.history['loss'])
plt.plot(history.history['val_loss'])
plt.title('Model Loss')
plt.ylabel('Loss')
plt.xlabel('Epoch')
plt.legend(['Train', 'Validation'], loc='upper left')

plt.tight_layout()
plt.show()
```

## SECTION 12 — Model Evaluation

This section evaluates the trained model's performance on the test set. It generates a confusion matrix heatmap to visualize the performance of the classification model, showing the counts of true-positive, true-negative, false-positive, and false-negative predictions. A classification report is also generated, providing precision, recall, f1-score, and support for each class.

```python
# Predict probabilities for the test set
y_pred_probs = model.predict(X_test)
y_pred = np.argmax(y_pred_probs, axis=1)
y_true = np.argmax(y_test, axis=1)

# Generate Confusion Matrix
cm = confusion_matrix(y_true, y_pred)
plt.figure(figsize=(8, 6))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', xticklabels=label_encoder.classes_, yticklabels=label_encoder.classes_)
plt.title('Confusion Matrix')
plt.xlabel('Predicted Label')
plt.ylabel('True Label')
plt.show()

# Generate Classification Report
report = classification_report(y_true, y_pred, target_names=label_encoder.classes_)
print("Classification Report:")
print(report)
```

## SECTION 13 — Prediction Visualization

This section visualizes the model's predictions on a few random MRI images from the test set. For each image, it displays the actual label, the predicted label, and the prediction confidence. This provides a qualitative assessment of the model's performance.

```python
num_samples = 5 # Number of random samples to display

plt.figure(figsize=(15, 10))

for i in range(num_samples):
    idx = np.random.randint(0, len(X_test))
    sample_image = X_test[idx]
    true_label_encoded = np.argmax(y_test[idx])
    true_label = label_encoder.inverse_transform([true_label_encoded])[0]

    # Make prediction
    prediction_probs = model.predict(np.expand_dims(sample_image, axis=0))[0]
    predicted_label_encoded = np.argmax(prediction_probs)
    predicted_label = label_encoder.inverse_transform([predicted_label_encoded])[0]
    confidence = prediction_probs[predicted_label_encoded] * 100

    plt.subplot(1, num_samples, i + 1)
    plt.imshow(sample_image)
    plt.title(f"True: {true_label}\nPred: {predicted_label}\nConf: {confidence:.2f}%")
    plt.axis('off')

plt.suptitle('Sample Predictions on Test Images', fontsize=16)
plt.tight_layout(rect=[0, 0.03, 1, 0.95])
plt.show()
```

## SECTION 14 — Explainability using Grad-CAM

This section implements Grad-CAM (Gradient-weighted Class Activation Mapping) to visualize the regions in the MRI image that are most important for the model's classification decision. Grad-CAM provides a heatmap that highlights discriminative regions, offering insight into the model's 'reasoning' for its predictions.

```python
def make_gradcam_heatmap(img_array, model, last_conv_layer_name, pred_index=None):
    # First, we create a model that maps the input image to the activations
    # of the last convolutional layer as well as the output predictions
    grad_model = tf.keras.models.Model(
        model.inputs, [model.get_layer(last_conv_layer_name).output, model.output]
    )

    # Then, we compute the gradient of the top predicted class for our input image
    # with respect to the activations of the last convolutional layer
    with tf.GradientTape() as tape:
        last_conv_layer_output, preds = grad_model(img_array)
        if pred_index is None:
            pred_index = tf.argmax(preds[0])
        class_channel = preds[:, pred_index]

    # This is the gradient of the output neuron (top predicted or chosen) with regard to
    # the output feature map of the last convolutional layer
    grads = tape.gradient(class_channel, last_conv_layer_output)

    # This is a vector where each entry is the mean intensity of the gradient
    # over a channel for a specific feature map entry
    pooled_grads = tf.reduce_mean(grads, axis=(0, 1, 2))

    # We multiply each channel in the feature map array by "how important" this channel is
    # with regard to the top predicted class, then sum all the channels to obtain
    # the heatmap class activation
    last_conv_layer_output = last_conv_layer_output[0]
    heatmap = last_conv_layer_output @ pooled_grads[..., tf.newaxis]
    heatmap = tf.squeeze(heatmap)

    # For visualization purpose, we will normalize the heatmap between 0 & 1
    heatmap = tf.maximum(heatmap, 0) / tf.math.reduce_max(heatmap)
    return heatmap.numpy()

def display_gradcam(img, heatmap, alpha=0.4):
    # Rescale heatmap to a range 0-255
    heatmap = np.uint8(255 * heatmap)

    # Use a colormap to colorize the heatmap
    jet = plt.colormaps['jet']

    # Use RGB values of the colormap
    jet_colors = jet(np.arange(256))[:, :3]
    jet_heatmap = jet_colors[heatmap]

    # Create an image with RGB colorized heatmap
    jet_heatmap = keras.utils.array_to_img(jet_heatmap)
    jet_heatmap = jet_heatmap.resize((img.shape[1], img.shape[0]))
    jet_heatmap = keras.utils.img_to_array(jet_heatmap)

    # Superimpose the heatmap on original image
    superimposed_img = jet_heatmap * alpha + img
    superimposed_img = keras.utils.array_to_img(superimposed_img)

    return superimposed_img

# --- Grad-CAM Visualization ---

# Find the name of the last convolutional layer in MobileNetV2
# MobileNetV2 structure usually has 'block_16_project_1_relu' as the last conv layer before pooling
last_conv_layer_name = None
for layer in reversed(model.layers):
    if isinstance(layer, (tf.keras.layers.Conv2D, tf.keras.layers.DepthwiseConv2D)) and 'conv' in layer.name:
        last_conv_layer_name = layer.name
        break

if last_conv_layer_name is None:
    print("Could not find a suitable last convolutional layer for Grad-CAM.")
else:
    print(f"Last convolutional layer found: {last_conv_layer_name}")

    # Select a random test image
    sample_idx = np.random.randint(0, len(X_test))
    test_image = X_test[sample_idx]
    true_label_encoded = np.argmax(y_test[sample_idx])
    true_label = label_encoder.inverse_transform([true_label_encoded])[0]

    # Prepare image for model prediction
    img_array = np.expand_dims(test_image, axis=0)

    # Predict the class for the image
    preds = model.predict(img_array)
    predicted_class_idx = np.argmax(preds[0])
    predicted_label = label_encoder.inverse_transform([predicted_class_idx])[0]

    # Generate heatmap
    heatmap = make_gradcam_heatmap(img_array, model, last_conv_layer_name, predicted_class_idx)

    # Display results
    plt.figure(figsize=(15, 5))

    plt.subplot(1, 3, 1)
    plt.imshow(test_image)
    plt.title(f'Original Image\nTrue: {true_label}\nPred: {predicted_label}')
    plt.axis('off')

    plt.subplot(1, 3, 2)
    plt.imshow(heatmap)
    plt.title('Grad-CAM Heatmap')
    plt.axis('off')

    superimposed_img = display_gradcam(test_image, heatmap)
    plt.subplot(1, 3, 3)
    plt.imshow(superimposed_img)
    plt.title('Overlayed with Heatmap')
    plt.axis('off')

    plt.suptitle('Grad-CAM Visualization for a Sample Test Image', fontsize=16)
    plt.tight_layout(rect=[0, 0.03, 1, 0.95])
    plt.show()
```
